```
import os 

print(os.listdir('.')) # List semua file di directory sekarang (.)

print(os.listdir('./sample_data')) # List file di directory sample_data

directory = os.path.join('sample_data', 'california_housing_train.csv') # Output: sample_data/california_housing_train.csv

text_file = open(directory, "r")

lines = text_file.readlines()

for line in lines[:3]: # Print 3 baris pertama
  split_columns = line.split(",")

parsed_values = []

for index, line in enumerate(lines):
  split_columns = line.split(",")

  if index == 0:
    continue
  
  parsed_values.append([float(split_columns[0]), float(split_columns[1]), float(split_columns[2]), float(split_columns[3]), float(split_columns[4]), float(split_columns[5]), float(split_columns[6]), float(split_columns[7]), float(split_columns[8])])

print("\n================ ASSINGMENT #4 ================ ")

from statistics import mean
import numpy as np

data1 = [] #housing_median_age
data2 = [] #total_rooms
data3 = [] #total_bedrooms
data4 = [] #population
data5 = [] #households
data6 = [] #median_income
data7 = [] #median_house_value

for i in parsed_values :
  data1.append(i[2])
  data2.append(i[3])
  data3.append(i[4])
  data4.append(i[5])
  data5.append(i[6]) 
  data6.append(i[7])
  data7.append(i[8])

print("\nMean \n") 
print("Rata-rata housing_median_age : ",mean(data1))
print("Rata-rata total_rooms        : ",mean(data2))
print("Rata-rata total_bedrooms     : ",mean(data3))
print("Rata-rata population         : ",mean(data4))
print("Rata-rata households         : ",mean(data5))
print("Rata-rata median_income      : ",mean(data6))
print("Rata-rata median_house_value : ",mean(data7))
print(" ")
print(" ")
print("Standard Deviation \n")
print("Standard deviasi housing_median_age : ",np.std(data1))
print("Standard deviasi total_rooms        : ",np.std(data2))
print("Standard deviasi total_bedrooms     : ",np.std(data3))
print("Standard deviasi population         : ",np.std(data4))
print("Standard deviasi households         : ",np.std(data5))
print("Standard deviasi median_income      : ",np.std(data6))
print("Standard deviasi median_house_value : ",np.std(data7))
print(" ")
print(" ")
print("Maximum\n")
print("Nilai maximum dari housing_median_age :",max(data1))
print("Nilai maximum dari total_rooms        :",max(data2))
print("Nilai maximum dari total_bedrooms     :",max(data3))
print("Nilai maximum dari population         :",max(data4))
print("Nilai maximum dari households         :",max(data5))
print("Nilai maximum dari median_income      :",max(data6))
print("Nilai maximum dari median_house_value :",max(data7))
print(" ")
print(" ")
print("Minimum\n")
print("Nilai minimum dari housing_median_age :",min(data1))
print("Nilai minimum dari total_rooms        :",min(data2))
print("Nilai minimum dari total_bedrooms     :",min(data3))
print("Nilai minimum dari population         :",min(data4))
print("Nilai minimum dari households         :",min(data5))
print("Nilai minimum dari median_income      :",min(data6))
print("Nilai minimum dari median_house_value :",min(data7))
print(" ")
print(" ")
print("Q1 (Kuartal Pertama) \n")
print("Nilai Q1 dari housing_median_age : ",np.quantile(data1, .25))
print("Nilai Q1 dari total_rooms        : ",np.quantile(data2, .25))
print("Nilai Q1 dari total_bedrooms     : ",np.quantile(data3, .25))
print("Nilai Q1 dari population         : ",np.quantile(data4, .25))
print("Nilai Q1 dari households         : ",np.quantile(data5, .25))
print("Nilai Q1 dari median_income      : ",np.quantile(data6, .25))
print("Nilai Q1 dari median_house_value : ",np.quantile(data7, .25))
print(" ")
print(" ")
print("Q2 (Kuartal Kedua) \n")
print("Nilai Q2 dari housing_median_age : ",np.quantile(data1, .50))
print("Nilai Q2 dari total_rooms        : ",np.quantile(data2, .50))
print("Nilai Q2 dari total_bedrooms     : ",np.quantile(data3, .50))
print("Nilai Q2 dari population         : ",np.quantile(data4, .50))
print("Nilai Q2 dari households         : ",np.quantile(data5, .50))
print("Nilai Q2 dari median_income      : ",np.quantile(data6, .50))
print("Nilai Q2 dari median_house_value : ",np.quantile(data7, .50))
print(" ")
print(" ")
print("Q3 (Kuartal Ketiga) \n")
print("Nilai Q3 dari housing_median_age : ",np.quantile(data1, .75))
print("Nilai Q3 dari total_rooms        : ",np.quantile(data2, .75))
print("Nilai Q3 dari total_bedrooms     : ",np.quantile(data3, .75))
print("Nilai Q3 dari population         : ",np.quantile(data4, .75))
print("Nilai Q3 dari households         : ",np.quantile(data5, .75))
print("Nilai Q3 dari median_income      : ",np.quantile(data6, .75))
print("Nilai Q3 dari median_house_value : ",np.quantile(data7, .75))
print(" ")
print(" ")
print("Apakah terdapat nilai kosong (0) ? \n")

for i in parsed_values :
  for value in i :
    if value == 0 :
      print("Terdapat angka kosong (0)") 
      break
    else : 
      print("Tidak terdapat angka kosong (0)")
      break
  break
```
