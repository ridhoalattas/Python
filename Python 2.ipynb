```
print("No. 1")
print(" ")

array_of_dictionaries = [{
    "name": "Budi",
    "age": 10,
    "test_scores": [100.0, 98.0, 89.0]
},
{
    "name": "Charlie",
    "age": 12,
    "test_scores": [90.0, 100.0]
}]

for value in array_of_dictionaries:
  for dictionary_list in value : 
    if dictionary_list == "test_scores" : 
      nama = value["name"]
      umur = value["age"]
      avg = round(sum(value['test_scores']) / len(value['test_scores']),2)
      print(f"Pelajar {str(nama)} dengan umur {str(umur)} mendapatkan nilai rata-rata {str(avg)}")
```

```
print("No. 2")
print(" ")

def get_student_min_score(array_list):
  for value in array_list:
    for dictionary_list in value :
      if dictionary_list == "test_scores" :
        nama = value['name']
        umur = value['age']
        min_score = min(value['test_scores'])
        print(f"Pelajar {str(nama)} dengan umur {str(umur)} mendapatkan nilai terkecil {str(min_score)}")

get_student_min_score(array_of_dictionaries)
```

```
print("No. 3")
print(" ")

def get_student_max_score(array_list):
  for value in array_list :
    for dictionary_list in value :
      if dictionary_list == 'test_scores' :
        nama = value['name']
        umur = value['age']
        max_score = max(value['test_scores'])
        print(f"Pelajar {str(nama)} dengan umur {str(umur)} mendapatkan nilai terbesar {str(max_score)}")

get_student_max_score(array_of_dictionaries) 
```
