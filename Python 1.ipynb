```python
print("""
1. Buatlah suatu list kosong bernama numbers. Lalu, buatlah loop yang secara otomatis menambahkan nomor 1-10 ke list tersebut. 
◦ Hint: Gunakan function range() untuk melakukan loop selama n kali
""")
numbers = []
for i in range(1, 11): 
    numbers.append(i)
print(numbers)

print("\n2. Tambahkan setiap nilai numbers sebanyak 2. \n")
numbers2 = []
for i in numbers:
    i += 2
    numbers2.append(i)
print(numbers2)

print("\n3. Cek untuk setiap angka ganjil, apabila ganjil tambahkan angka 1.\n")
numbers3 = []
for i in numbers2 :
    if i % 2 != 0:
        i += 1
        numbers3.append(i)
    elif i % 2 == 0:
        numbers3.append(i)
print(numbers3)

print("\n4. Hapus semua angka yang lebih kecil dari 5 di list.\n")

while(numbers3.count(4)):
    numbers3.remove(4)
print(numbers3)

print("""\n
5. Print semua nilai isi di list dengan format:
◦ Angka ke-(index) bernilai (angka)
◦ Misal: Angka ke-0 bernilai 10
""")

for index, numbers5 in enumerate(numbers3):
    print(f"Angka ke-{index} : {numbers5}")

print("""\n
6. Print semua angka yang habis dibagi 2, dengan format: 
◦ Angka ke-(index) bernilai (angka)
◦ Misal: Angka ke-0 bernilai 10 
""")
for index, numbers6 in enumerate(numbers3) : 
    if numbers6 % 2 == 0 :
        print(f"Angka ke-{index} : {numbers6}")
```
